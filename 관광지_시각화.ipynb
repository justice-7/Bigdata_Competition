```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/관광.csv')
df.head()
```

```python
df.drop('총계', axis=1, errors='ignore', inplace=True)
df =df.transpose()
df=df.rename(columns=df.iloc[0])
df.drop(df.index[0], inplace=True)
df.head()
```

```python
df.info()
```

```python
# 형변환 str->float
for i in df.columns:
    df[i] = df[i].str.replace(',','')
    df[i]=df[i].astype(float)
```

```python
# 결측치 0으로 채움
df=df.fillna(0)
df.head()
```
